# Exploratory Data Analysis on test set

```python
import pandas as pd
```

```python
#import test_set
df_validate=pd.read_csv('./data/test_set.csv')
```

```python
#import plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_validate['review_count'].hist()
plt.title('Histogram of Review Count')
plt.xlabel('Review Count')
plt.ylabel('Frequency')
plt.show()
```

```python
#histogram of ratings
df_validate['rating'].hist()
plt.title('Histogram of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Showing the data on the map
df_validate.plot.scatter(x='longitude', y='latitude')
plt.title('Scatter Plot of Latitude vs Longitude')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

```python
# See if there are any correlations
import seaborn as sns
import numpy as np

# Select only numeric columns
numeric_cols = df_validate.select_dtypes(include=np.number)

# Calculate correlation matrix
corr = numeric_cols.corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
